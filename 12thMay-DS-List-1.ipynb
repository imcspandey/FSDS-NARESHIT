```python
l = [1]
l.append('2')
l.append(1)
l.append(2+3j)
l.append([9,8,7])
print(l)
print("is '2' in l: ", '2' in l)
print(type(l))
l2 = l
l3 = l.copy()
print("ID of l is equal to l2: ", id(l) == id(l2))
print("ID of l is equal to l3: ", id(l) == id(l3))

l2.clear() # it will clear l2 and l both since l2 has reference of l
print("after clearing l and l2: ", l, l2)
print("l3: ", l3)

print("count of 1 in list: ", l3.count(1))

print(l3[::1])
print(l3[::-1])
print(l3[::100])
print(l3[::-2])
print(l3[1::-2])
```
