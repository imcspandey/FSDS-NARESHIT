```python
l3 = [1, '2', 1, (2+3j), [9, 8, 7]]
l3[0] = "5"
print(l3)

l3[-1] = [3,4,5]
print(l3)
print("nested indexing: ", l3[-1][-1])

l3.extend([6,7,8])
print("l3 after extending: ", l3)

l3.pop() # defaults index is -1. can pass desired index too
print("l3 after popping. removes last element: ", l3)

l3.insert(2, 23)
print("l3 after inserting 23 at index 2: ", l3)

l3.pop(2)
print("l3 after popping of index 2. removes element at index 2: ", l3)

l3.remove('5')
print("l3 after removing element '5': ", l3)

l3.reverse()
print("l3 after reversing: ", l3)

l4 = [5,6,1,7,0,9,8]
print(l4)
l4.sort()
print("l4 after doing sorting. sorts in ascending order by default: ", l4)

l4.sort(reverse = True)
print("l4 after doing sorting. sorts in descending order: ", l4)

print("printing l4 elements using loop-1:")
for n in l4:
    print(n)

print("printing l4 elements using loop-2(enumerate):")
for n in enumerate(l4):
    print(n)

print("all function : ", all(l4)) # False because of one falsy value

l4.remove(0)
print("l4 after removing falsy value: ", l4)

print("all function now: ", all(l4)) # True because no falsy value

print("any function: ", any(l4)) # True because all are Truthy value

l5 = [None, '', 0, []]
print("l5 with Falsy values: ", l5)
print("any function now:", any(l5)) 
```
