```python
p = 20
q = 20
r = q
print(f"IDs p, q, r: {id(p)}, {id(q)}, {id(r)}")
print(f"ID HEXs p, q, r: {hex(id(p))}, {hex(id(q))}, {hex(id(r))}")
print(id(r) == id(q))
print(hex(id(q)) == hex(id(p)))
```
