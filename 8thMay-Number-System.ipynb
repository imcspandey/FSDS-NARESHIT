```python
# BINARY NUMBER SYSTEM
# It has 0b prefix. keep dividing the number by 2 until not dividable.
```

```python
bin(2) # output: 10
```

```python
bin(0) # output: 0
```

```python
bin(14) # output: 1110
```

```python
int(0b10) # output: 2
```

```python
int(0b110) # output: 6
```

```python
bin(True) # output: 1
```

```python
bin(False) # output: 0
```

```python
# OCTAL NUMBER SYSTEM
# It has 0o prefix. keep dividing the number by 8 until not dividable.
```

```python
oct(45) # output: 55
```

```python
int(0o55) # output: 45
```

```python
oct(True) # output: 1
```

```python
oct(False) # output: 0
```

```python
# HEXADECIMAL NUMBER SYSTEM
# It has 0x prefix. keep dividing the number by 16 until not dividable.
```

```python
hex(17) # output: 11
```

```python
hex(171) # output: ab
```

```python
hex(True) # output: 1
```

```python
hex(False) # output: 0
```

```python
# BITWISE OPERATOR
"""
1. Complement (~)
2. Bitwise and (&)
3. Bitwise or (|)
4. Exclusive or (xor, ^)
5. Left shift (<<)
6. Right shift (>>)
"""
```

```python
# COMPLEMENT OPERATOR
# It flips all the bits
```

```python
~123 # output: -124
```

```python
~ -123
```

```python
~567 # output: -568
```

```python
# BITWISE AND (&)
# it returns 1 only when bits are 1
```

```python
print(67 & 45) # output: 1
print(bin(67))
print(bin(45))
```

```python
# BITWISE OR (|)
# it returns 1 if one of the bit is 1
```

```python
85 | 56 # output 125
```

```python
print(bin(85))
print(bin(56))
```

```python
# XOR OPERATOR (^)
# It returns 1 only when bits are different.
```

```python
13^12 # output: 1
```

```python
56 ^ 45 # output: 21T
```

```python
# LEFT SHIFT OPERATOR
# it fills zero from right
```

```python
3 << 2 # output: 12
```

```python
15 << 3 # output: 120
```

```python
# RIGHT SHIFT OPERATOR
# It removes bits from right
```

```python
97 >> 2 # output: 24
```

```python
97 >> 3 # output: 12
```
